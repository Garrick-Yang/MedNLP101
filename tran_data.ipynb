```python
import json
import random
```

```python
#本块为CHIP-CDN

#输入路径
with open('pre-datasets/CHIP-CDN/CHIP-CDN_test.json', 'r') as file_CDN:
    data_CDN = json.load(file_CDN)

input_CDN = [
    "给出下面诊断原词的标准化：\\n[INPUT_TEXT]\\n候选集：[LIST_LABELS]\\n说明：从候选的若干个ICD-10诊断标准词中选择出与原诊断描述匹配的词\\n答：",
    "找出归一后的标准词：\\n[INPUT_TEXT]\\n选项：[LIST_LABELS]\\n说明：从候选的若干个ICD-10诊断标准词中选择出与原诊断描述匹配的词\\n答：",
    "诊断归一化：\\n[INPUT_TEXT]\\n选项：[LIST_LABELS]\\n说明：从候选的若干个ICD-10诊断标准词中选择出与原诊断描述匹配的词\\n答：",
    "诊断实体的语义标准化：\\n[INPUT_TEXT]\\n实体选项：[LIST_LABELS]\\n说明：从候选的若干个ICD-10诊断标准词中选择出与原诊断描述匹配的词\\n答：",
    "给出诊断的归一化：\\n[INPUT_TEXT]\\n医学实体选项：[LIST_LABELS]\\n说明：从候选的若干个ICD-10诊断标准词中选择出与原诊断描述匹配的词\\n答：",
    "[INPUT_TEXT]\\n归一化后的标准词是？\\n实体选项：[LIST_LABELS]\\n说明：从候选的若干个ICD-10诊断标准词中选择出与原诊断描述匹配的词\\n答：",
    "实体归一化：\\n[INPUT_TEXT]\\n实体候选：[LIST_LABELS]\\n说明：从候选的若干个ICD-10诊断标准词中选择出与原诊断描述匹配的词\\n答："
]

n = 0
list_CDN = []
for item in data_CDN:
    n = n + 1
    input_template = random.choice(input_CDN)
    format_item = {
        #"input": input_template.replace("[INPUT_TEXT]", item["text"]).replace("[LIST_LABELS]", "，".join(item["normalized_result"].split("##"))),
        "input": input_template.replace("[INPUT_TEXT]", item["text"]).replace("[LIST_LABELS]", ""),
        "target": "，".join(item["normalized_result"].split("##")),
        "answer_choices": [""],
        "task_type": "normalization",
        "task_dataset": "CHIP-CDN",
        "sample_id": n
    }
    list_CDN.append(format_item)

#输出路径
with open('datasets/CHIP-CDN/formatted_test.json', 'w') as output_CDN:
    for format_item in list_CDN:
        json.dump(format_item, output_CDN, ensure_ascii=False)
        output_CDN.write('\n')  
```

```python
#本块为CHIP-STS

#输入路径
with open('pre-datasets/CHIP-STS/CHIP-STS_train.json', 'r') as file_STS:
    data_STS= json.load(file_STS)

#请修改，label=1为train和dev，label=0为test
label_STS = 1

input_STS = [
    "以下两句话的意思相同的吗？\\n“[INPUT_TEXT_1]”，“[INPUT_TEXT_2]”\\n选项：是的，不是\\n答：",
    "我想知道下面两句话的意思是否相同。\\n“[INPUT_TEXT_1]”，“[INPUT_TEXT_2]”。\\n选项：是的，不是\\n答：",
    "我是否可以用以下的句子：“[INPUT_TEXT_1]”，来替换这个句子：“[INPUT_TEXT_2]”，并且它们有相同的意思？\\n选项：是的，不是\\n答：",
    "“[INPUT_TEXT_1]”和“[INPUT_TEXT_2]”是同一个意思吗？\\n选项：是的，不是\\n答：",
    "“[INPUT_TEXT_1]”，“[INPUT_TEXT_2]”。\\n这两句是一样的意思吗？\\n选项：是的，不是\\n答："
  ]

map_STS = {
    "1":"是的",
    "0":"不是"
}

list_STS = []
for item in data_STS:
    input_template = random.choice(input_STS)
    format_item = {
        "input": input_template.replace("[INPUT_TEXT_1]", item["text1"]).replace("[INPUT_TEXT_2]", item["text2"]),
        "target":"" if label_STS==0 else map_STS[item["label"]],
        "answer_choices": ["相同", "不同"],
        "task_type": "cls",
        "task_dataset": "CHIP-STS",
        "sample_id": item["id"]
    }
    list_STS.append(format_item)

#输出路径
with open('datasets/CHIP-STS/formatted_train.json', 'w') as output_STS:
    for format_item in list_STS:
        json.dump(format_item, output_STS, ensure_ascii=False)
        output_STS.write('\n') 
```

```python
#本块为KUAKE-QTR

#输入路径
with open('PromptCBLUE-main/pre-datasets/KUAKE-QTR/KUAKE-QTR_train.json', 'r') as file_QTR:
    data_QTR = json.load(file_QTR)

#请修改，label=1为train和dev，label=0为test
label_QTR = 1   

input_QTR = [
    "以下两句话的意思相同的吗？\\n“[INPUT_TEXT_1]”，“[INPUT_TEXT_2]”。\\n选项：[LIST_LABELS]\\n答：",
    "我想知道下面两句话的意思有多相似。\\n“[INPUT_TEXT_1]”，“[INPUT_TEXT_2]”\\n选项：[LIST_LABELS]\\n答：",
    "下面的搜索词和页面标签的意思有多相同？\\n搜索词：[INPUT_TEXT_1]\\n页面标签：[INPUT_TEXT_2]\\n选项：[LIST_LABELS]\\n答：",
    "下面两个句子的语义相似程度是[LIST_LABELS]中的哪一种？\\n“[INPUT_TEXT_1]”，“[INPUT_TEXT_2]”\\n答：",
    "“[INPUT_TEXT_1]”和“[INPUT_TEXT_2]”是同一个意思吗？\\n选项：[LIST_LABELS]\\n答：",
    "“[INPUT_TEXT_1]”和“[INPUT_TEXT_2]”的意思有多相似？\\n选项：[LIST_LABELS]\\n答：",
    "“[INPUT_TEXT_1]”，“[INPUT_TEXT_2]”。\\n这两句话的意思的匹配程度如何？\\n选项：[LIST_LABELS]\\n答：",
    "搜索词：“[INPUT_TEXT_1]”。页面标题：“[INPUT_TEXT_2]”。这两句是一样的意思吗？选项：[LIST_LABELS]\\n答："
]

map_QTR = {
    "0":"完全不匹配或者没有参考价值",
    "1":"很少匹配有一些参考价值",
    "2":"部分匹配",
    "3":"完全匹配"
}

list_QTR = []
for item in data_QTR:
    input_template = random.choice(input_QTR)
    format_item = {
        "input": input_template.replace("[INPUT_TEXT_1]", item["query"]).replace("[INPUT_TEXT_2]", item["title"]).replace("[LIST_LABELS]",",".join(map_QTR.values())),
        "target": "" if label_QTR==0 else map_QTR[item["label"]],
        "answer_choices": ["完全不匹配或者没有参考价值", "很少匹配有一些参考价值", "部分匹配", "完全匹配"],
        "task_type": "matching",
        "task_dataset": "KUAKE-QTR",
        "sample_id": item["id"]
    }
    list_QTR.append(format_item)


#输出路径
with open('PromptCBLUE-main/datasets/KUAKE-QTR/formatted_train.json', 'w') as output_QTR:
    for format_item in list_QTR:
        json.dump(format_item, output_QTR, ensure_ascii=False)
        output_QTR.write('\n')  
```
